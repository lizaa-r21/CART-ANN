```python
listcolumns= ['spending', 'advance_payments', 'probability_of_full_payment',
       'current_balance', 'credit_limit', 'min_payment_amt',
       'max_spent_in_single_shopping']
def univaranalysis(column,nbins):
  print('Description of'+ column)
  print('------------------------')
  print(df[column].describe(), end='')
  print('')
  print('')
  
  print('Ranges of'+ column)
  print('-------------------')
  print(df[column].max()-df[column].min(), end='')
  print('')
  print('')

  plt.figure
  print('Distribution of'+ column)
  print('------------------------')
  sns.histplot(df[column], kde= True)
  plt.show()
  
  plt.figure
  print('boxplot of'+ column)
  print('-------------------')
  sns.boxplot(df[column], notch= True)
  plt.show()
```

```python
#kmean clustering
```

```python
from sklearn.cluster import KMeans 
```

```python
wss =[]
for i in range(1,11):
    KM = KMeans(n_clusters=i)
    KM.fit(df1)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
plt.grid()

```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
k_means = KMeans(n_clusters = 2)
```

```python
k_means.fit(df1)
```

```python
labels = k_means.labels_
```

```python
silhouette_score(df1, labels)
```

```python
silhouette_samples(df1, labels).min()
```
